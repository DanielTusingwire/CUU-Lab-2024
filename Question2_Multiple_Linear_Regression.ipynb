a) Produce a multiple Linear regression graph regards Serious, Fatal and Minor. (10 marks)

b) Determine the coefficients and intercept values of your graph. (5 marks)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/drive/MyDrive/SIMULATION_FINAL/Risk_Factors.csv")
data
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data)
df
```

```python
# extracting the predictor variables and the target variable

X = df[['Serious', 'Fatal', 'Minor']]  # Independent variables
y = df['Total_Crashes']
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
```

```python
import seaborn as sns

width = 10
height = 5
plt.figure(figsize=(width, height))

ax1 = sns.distplot(y, hist=False, color='r', label='Actual values')
sns.distplot(y_pred, hist=False, color='g', label='Predicted values', ax=ax1)

plt.title('Actual values vs Predicted values for Total Crushes')
plt.xlabel('Actual Values (Serious, Fatal, Minor)')
plt.ylabel('predicted values for Total Crushes')
plt.legend()


plt.show()
plt.close()
```

# **b) Determine the coefficients and intercept values of your graph**

```python
intercept = model.intercept_
coeff = model.coef_

print("Intercept:", intercept)
print("Coefficients :", coeff)
```

# **c) State the multiple linear regression model mathematically. (5 marks)**

# **d) Evaluate the model by obtaining the values of R-squared and mean squared errors**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Model performance on training data
r_squared = r2_score(y, y_pred)

mse = mean_squared_error(y, y_pred)
print("\nModel Performance Metrics:")
print("R-squared:", r_squared)
print("Mean Squared Error:", mse)
```
